```python
# Load the GPT-2 language model and tokenizer
gpt2_model_name = "gpt2"  # Use the smaller GPT-2 variant
tokenizer = GPT2Tokenizer.from_pretrained(gpt2_model_name)
model = GPT2LMHeadModel.from_pretrained(gpt2_model_name)
```

```python
# Load the CLIP model and processor
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch16")
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch16")
```

```python
from PIL import Image
# Generate a prompt using GPT-2
prompt_text = "Generate a description of an image showing"
input_ids = tokenizer.encode(prompt_text, return_tensors="pt")

# Load the image using PIL (replace "/path/to/your/image.jpg" with the actual path)
image_path = "/content/Buildings.jpg"
image = Image.open(image_path)

# Process the image and prompt with CLIP
inputs = clip_processor(prompt_text, images=image, return_tensors="pt", padding=True)

with torch.no_grad():
    logits_per_image = clip_model(**inputs).logits_per_image
    probs = logits_per_image.softmax(dim=1)[0]

# Manually find the predicted class label
class_labels = ['buildings', 'forest', 'glacier', 'mountain', 'sea', 'street']
predicted_label_idx = torch.argmax(probs).item()
predicted_label = class_labels[predicted_label_idx]

# Print the CLIP-based classification result
print("CLIP Predicted Class:", predicted_label)
```
