```python
from PIL import Image
import requests
from transformers import CLIPModel, CLIPProcessor
import torch
import numpy as np
```

```python
model = CLIPModel.from_pretrained("openai/clip-vit-large-patch14")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-large-patch14")
```

```python
image_path = "/content/landscape.jpg"
image = Image.open(image_path)
text = ["A photo of calm landscape", "A picture of dogs and cats", "A photo of blue horses and white houses"]
inputs = processor(text, images=image, return_tensors="pt", padding=True)
```

```python
with torch.no_grad():
  output = model(**inputs)

  text_features = output.text_embeds
  image_features = output.image_embeds

logits_per_image = output.logits_per_image

prob = logits_per_image.softmax(dim=-1).cpu().numpy()

print(prob)
```
